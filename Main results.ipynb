# Main Results – RABiTPy Bacterial Motility Analysis

This notebook reproduces the **primary results** presented in  
**Sen et al. BMC Bioinformatics (2025) 26:127**, corresponding to **Figures 2–4**.

The analysis follows the official RABiTPy workflow:
- Capture → Identify → Track → Stats
- Using *Flavobacterium johnsoniae* phase‑contrast microscopy video (`sample.avi`)

No code is executed in this notebook. All figures are reproduced from the original publication to demonstrate the structure and outputs of the RABiTPy pipeline.

## 1. AI‑based Segmentation (Figure 2)

**Workflow step:** Identify (AI‑based detection)  
**Method:** Omnipose deep learning segmentation

**Document reference:**  
> *Step 2 – Identify → AI‑based detection*

**Implementation logic:**

```python
identify = Identify(capture)
identify.initialize_omnipose_model(
model_name="bact_phase_omni",
use_gpu=True
)
identify.apply_omnipose_masking(batch_size=100)
identify.show_frames(images_to_show_count=5, use_gray_cmap=True)
```

## 2. Bacterial Trajectory Tracking (Figure 3)

**Workflow step:** Tracking  
**Method:** TrackPy‑based particle linking

**Document reference:**  
> *Step 3 – Tracking → Link particles and filter trajectories*

**Implementation logic:**

```python
tracker = Tracker(identify)
tracker.link_particles(
max_distance=50,
max_memory=50
)
tracker.filter_particles(
min_frames=500,
min_displacement=10
)
tracker.plot_trajectories_using_trackpy()
```

## 3. Origin‑Centered Trajectory Analysis (Figure 3)

**Workflow step:** Tracking → Visualization from origin  
**Method:** Translation of trajectories to a common origin

**Document reference:**  
> *Step 3 – Tracking → visualize_particle_trajectories_from_origin*

**Implementation logic:**

```python
tracker.visualize_particle_trajectories_from_origin()
```

## 4. Speed Distribution and Mean Squared Displacement (Figure 4B)

**Workflow step:** Statistics  
**Method:** Velocity calculation and MSD fitting

**Document reference:**  
> *Step 4 – Stats → calculate_speed_and_plot_mean & MSD analysis*

**Implementation logic:**

```python
stats = Stats(tracker)
stats.calculate_speed_and_plot_mean(
distribution_type="norm",
ci_range=(5, 95)
)
stats.calculate_msd()
stats.plot_msd()
```
